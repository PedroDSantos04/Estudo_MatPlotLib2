```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('IceCreamData.csv')
df
```

```python
df.plot()
```

```python
valor = df.sort_values('Temperature')
valor
```

```python
plt.title('Ice Cream Truck')

plt.plot(valor['Temperature'], valor['Revenue' ], 'b--')
         
plt.ylabel('U$')
plt.xlabel('Temperatura')

plt.show()
```
